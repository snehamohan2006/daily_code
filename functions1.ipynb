```python
def calculate_grade(marks):
    total=sum(marks)
    average=total/len(marks)

    if average>=90:
        grade="A"
    elif average>=75:
        grade="B"
    elif average>=60:
        grade="C"

    else:
        grade="fail"

    return grade
marks=[90,90,95,90,90]
result=calculate_grade(marks)
print("grade",result)
```
