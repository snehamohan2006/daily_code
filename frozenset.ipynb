```python
a=frozenset({1,2,3,4,5})
b=frozenset({8,7,3,4})
print(a.difference(b))
print(a-b)

```

```python
a=frozenset({1,2,3,4,5})
b=frozenset({8,7,3,4})
print(a.intersection(b))
print(a & b)
```

```python
a=frozenset({1,2})
b=frozenset({3,4})
c=frozenset({3,6})
print(a.isdisjoint(b))
print(b.isdisjoint(c))
```
