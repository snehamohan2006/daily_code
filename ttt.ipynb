```python
n=int(input("enter a number:"))
original=n
sum_factorial=0

while n>0:
    digit=n%10
    fact=1
    for i in range(1,digit+1):
        fact=fact*i
    sum_factorial=sum_factorial+fact
    n=n//10
if sum_factorial==original:
    print("strong number")
else:
    print("not strong number")
    
      
```
