```python
def print_numbers(nums):
    for n in nums:
        print(n)

print_numbers([1,2,3,4])
```

```python
def print_list(lst):
    for item in lst:
        print(item)

numbers=[10,20,30,40]
print_list(numbers)
```

```python
def count_even(lst):
    count=0
    for num in lst:
        if num%2==0:
            count+=1
        return count
print(count_even([1, 2, 4, 7, 8]))  # Output: 3
```
