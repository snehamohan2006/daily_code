```python
def simple_interest(p, t, r=5):
    si = (p * t * r) / 100
    return si


# function call
result = simple_interest(1000, 2)
print("Simple Interest:", result)

# function call with custom rate
result2 = simple_interest(1000, 2, 10)
print("Simple Interest with custom rate:", result2)
```

```python
def sum_of_digits(n):
    total=0
    while n>0:
        digit=n%10
        total=total+digit
        n=n//10
    return total

number=1234
result=sum_of_digits(number)

print("sum of digits",result)
    


```
