```python
n=int(input("enter the number:"))

even=0
odd=0
zero=0

while n>0:
    digit=n%10
    
    if digit==0:
        zero=zero+1
    elif digit%2==0:
        even=even+1
    else:
        odd=odd+1
    n=n//10

print("number of even",even)
print("number of odd",odd)
print("number of zeroes",zero)

    

```
